```python
import json
import importlib
import re

# Notebook to check
notebook_path = "Pipeline-LinearSVC.ipynb"

# Read the notebook
with open(notebook_path, "r", encoding="utf-8") as f:
    nb = json.load(f)

modules = set()
import_pattern = re.compile(r'^\s*(?:import|from)\s+([\w\.]+)')

# Extract imports from code cells
for cell in nb['cells']:
    if cell['cell_type'] == 'code':
        for line in cell['source']:
            match = import_pattern.match(line)
            if match:
                modules.add(match.group(1).split('.')[0])

# Print version of each module
for module in sorted(modules):
    try:
        mod = importlib.import_module(module)
        version = getattr(mod, '__version__', 'no version info')
        print(f"{module}: {version}")
    except ImportError:
        print(f"{module}: not installed")

```
